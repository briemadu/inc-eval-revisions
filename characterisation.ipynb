# Characterisation of Revisions and Edits

In this notebook, we provide more examples for each type of edit and revision described in the paper.

```python
from inceval.incoutputs import IncOutputs
from inceval.aux import GOLD, SILVER
```

## Edits

### Convenience

Convenient edits change an incorrect label, whereas inconvenient edits change a correct label (into something wrong). 

The following edits are all inconvenient:

```python
outputs = IncOutputs(5, gold=[1,2,1,3,2], eval_mode=GOLD)
outputs.add_prefix(0, [1])
outputs.add_prefix(1, [1, 2])
outputs.add_prefix(2, [1, 1, 1])
outputs.add_prefix(3, [3, 1, 1, 3])
outputs.add_prefix(4, [3, 3, 1, 2, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```

The following edits are all convenient:

```python
outputs = IncOutputs(5, gold=[3,3,1,2,2], eval_mode=GOLD)
outputs.add_prefix(0, [1])
outputs.add_prefix(1, [1, 2])
outputs.add_prefix(2, [1, 1, 2])
outputs.add_prefix(3, [2, 2, 1, 3])
outputs.add_prefix(4, [3, 3, 1, 1, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```

### Effectiveness

Effective edits change a wrong label into a right one. Ineffective edtis turn a wrong label into another wrong label. Defective edits turn a wrong label into another wrong label.

The following edits are all effective:

```python
outputs = IncOutputs(5, gold=[3,3,1,2,2], eval_mode=GOLD)
outputs.add_prefix(0, [1])
outputs.add_prefix(1, [1, 2])
outputs.add_prefix(2, [1, 3, 2])
outputs.add_prefix(3, [3, 3, 1, 3])
outputs.add_prefix(4, [3, 3, 1, 2, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```

The following edits are all ineffective:

```python
outputs = IncOutputs(5, gold=[3,3,1,2,2], eval_mode=GOLD)
outputs.add_prefix(0, [1])
outputs.add_prefix(1, [1, 2])
outputs.add_prefix(2, [1, 1, 2])
outputs.add_prefix(3, [2, 2, 3, 3])
outputs.add_prefix(4, [1, 2, 3, 3, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```

The following edits are all defective:

```python
outputs = IncOutputs(5, gold=[3,3,1,2,2], eval_mode=GOLD)
outputs.add_prefix(0, [3])
outputs.add_prefix(1, [3, 3])
outputs.add_prefix(2, [1, 3, 1])
outputs.add_prefix(3, [1, 2, 1, 3])
outputs.add_prefix(4, [1, 2, 3, 3, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```

### Novelty

Novel edits change the label into a state it never was assigned before. Repetitive edits change the label into a state it already had before at some point.

The following edits are all innovative:

```python
outputs = IncOutputs(5, gold=[3,3,1,2,2], eval_mode=GOLD)
outputs.add_prefix(0, [3])
outputs.add_prefix(1, [2, 3])
outputs.add_prefix(2, [1, 3, 1])
outputs.add_prefix(3, [1, 2, 1, 3])
outputs.add_prefix(4, [4, 2, 3, 1, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```

```python
outputs = IncOutputs(5, gold=[3,3,1,2,2], eval_mode=GOLD)
outputs.add_prefix(0, [3])
outputs.add_prefix(1, [2, 1])
outputs.add_prefix(2, [1, 3, 1])
outputs.add_prefix(3, [2, 1, 1, 1])
outputs.add_prefix(4, [1, 3, 1, 1, 1])

outputs.plot_inc_chart(tokens=['this', 'is', 'an', 'example', '.'], figsize=(2, 2))
```
